```python
# 单变量
sns.set_style('darkgrid')
plt.figure(figsize=(10, 10))

def plot(index, column):
    plt.subplot(3, 1, index)
    sns.distplot(data[column])
    plt.ylabel(column)

columns = ['SALARY_MILLIONS', 'RPM', "AGE"]
for i in range(3):
    plot(i+1, columns[i])
```

```python
#双变量
```

```python
# 多变量
```

```python
multi_data = data.loc[:, ['RPM', 'SALARY_MILLIONS', 'AGE', 'POINTS']]
sns.pairplot(multi_data)
```

```python
# 衍生变量的一些可视化实践-以年龄为例
```

```python
def age_cut(x):
    """年龄划分"""
    if x <= 24:
        return "young"
    elif x >= 30:
        return "old"
    else:
        return "best"
data["age_cut"] = data.AGE.apply(age_cut)
data
```

```python
# 基于年龄段对球员薪水和时效值进行分析
sns.set_style('darkgrid')
plt.figure(figsize=(10, 10), dpi=100)
plt.title('RPM and Salary')

def plot2(feature, mark):
    x = data.loc[data.age_cut == feature].SALARY_MILLIONS
    y = data.loc[data.age_cut == feature].RPM
    plt.plot(x, y, mark)
plot2('old', '^')
plot2('best', 'x')
plot2('young', '.')
```

```python
multi_data2 = data.loc[:, ['RPM', 'POINTS', 'TRB', 'AST', 'STL', 'BLK', 'age_cut']]
sns.pairplot(multi_data2, hue='age_cut')
```

```python
## 球员数据分析
```

```python
# 球队薪资排行
salary = data.groupby('age_cut').agg({'SALARY_MILLIONS': np.max})
salary
```

```python
salary2 = data.groupby('age_cut', as_index=False).agg({'SALARY_MILLIONS': np.max})
salary2
```

```python
salary2.index
```

```python
data_team = data.groupby('TEAM')[['SALARY_MILLIONS']].mean()
data_team.sort_values(by="SALARY_MILLIONS", ascending=False)
```

```python
# 按照分球队分年龄段，上榜球员降序排列，如上榜球员数相同，则按效率值降序排列。
data.groupby(['TEAM', "age_cut"]).agg({"SALARY_MILLIONS": np.mean, "RPM": np.mean, "PLAYER": np.size})
```

```python
# 按球队综合实力排名
```

```python
data_rpm2 = data.groupby('TEAM', as_index=False).agg({"SALARY_MILLIONS": np.mean,
                                                      "RPM": np.mean,
                                                      "PLAYER": np.size,
                                                      "POINTS": np.mean,
                                                      "eFG%": np.mean,
                                                      "MPG": np.mean,
                                                      "AGE": np.mean})
data_rpm2
```

```python
data_rpm2.sort_values("RPM", ascending=False)
```

```python
# 利用箱线图和小提琴图进行数据分析
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20, 10))
data_team2 = data[data.TEAM.isin(['GS', 'CLE', 'SA', 'LAC', 'OKC', 'UTAH', 'CHA', 'TOR', 'NO', 'BOS'])]

# 进行相应的绘图
plt.subplot(3,1,1)
sns.boxplot(x="TEAM", y="SALARY_MILLIONS", data = data_team2)

plt.subplot(3,1,2)
sns.boxplot(x="TEAM", y="AGE", data = data_team2)

plt.subplot(3,1,3)
sns.boxplot(x="TEAM", y="MPG", data = data_team2)
```

```python
# 绘制小提琴图
sns.set_style("whitegrid")
plt.figure(figsize=(20, 10))

plt.subplot(3,1,1)
sns.violinplot(x="TEAM", y="3P%", data=data_team2)

plt.subplot(3,1,2)
sns.violinplot(x="TEAM", y="eFG%", data=data_team2)

plt.subplot(3,1,3)
sns.violinplot(x="TEAM", y="POINTS", data=data_team2)
```
