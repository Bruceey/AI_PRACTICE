```python
import random

x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
plt.figure(figsize=(25, 10), dpi=100)
plt.plot(x, y_shanghai)

#解决中文问题
plt.rcParams['font.sans-serif']=['Alibaba PuHuiTi']        # 用来显示中文标签
# plt.rcParams['axes.unicode_minus']=False   # 用来显示负号

#添加x，y刻度
x_ticks_label = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)
plt.xticks(x[::5], x_ticks_label)
plt.yticks(y_ticks[::5])

#添加网格
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel('时间', fontsize=20)
plt.ylabel('温度', fontsize=20)
plt.title("中午11点--12点某城市温度变化图", fontsize=20)
```

```python
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1,3) for i in x]

plt.figure(figsize=(25, 10), dpi=100)
plt.plot(x, y_shanghai, label="上海")
plt.plot(x, y_beijing, color='r', linestyle='--', label="北京")


#解决中文问题
plt.rcParams['font.sans-serif']=['Alibaba PuHuiTi']        # 用来显示中文标签
# plt.rcParams['axes.unicode_minus']=False   # 用来显示负号

#添加x，y刻度
x_ticks_label = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)
plt.xticks(x[::5], x_ticks_label)
plt.yticks(y_ticks[::5])

#添加网格
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc=0)
plt.xlabel('时间', fontsize=20)
plt.ylabel('温度', fontsize=20)
plt.title("中午11点--12点某城市温度变化图", fontsize=20)
```

```python
#利用subplots画图
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1,5) for i in x]

#创建画布
fig, axes = plt.subplots(1, 2, figsize=(20, 8), dpi=100)

axes[0].plot(x, y_shanghai, label="上海")
axes[1].plot(x, y_beijing, color='r', linestyle='--', label='北京')

x_ticks_label = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)

for axis in axes:
    axis.set_xticks(x[::5])
    axis.set_yticks(y_ticks[::5])
    axis.set_xticklabels(x_ticks_label[::5])
    axis.grid(True, linestyle='--', alpha=0.5)
    #添加描述信息
    axis.set_xlabel("时间")
    axis.set_ylabel("温度")
    axis.set_title("中午11点--12点某城市温度变化图", fontsize=20)
    axis.legend(loc=0)
plt.savefig("./上海和北京温度变化图.png")
```
