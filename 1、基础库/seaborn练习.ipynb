```python
# 绘制双变量
df = pd.DataFrame({'x': np.random.randn(500), 'y': np.random.randn(500)})
sns.jointplot('x', 'y', data=df)
```

```python
sns.jointplot('x', 'y', data=df, kind='hex')
```

```python
sns.jointplot('x', 'y', data=df, kind='kde')
```

```python
from sklearn.datasets import load_iris
data = load_iris()
data
```

```python
flower = pd.DataFrame(data.data, columns=data.feature_names)
flower['species'] = [data.target_names[i] for i in data.target]
flower
```

```python
sns.pairplot(flower)
```
